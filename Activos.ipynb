```python
import pandas as pd #pip install pandas
import yfinance as yf #pip install yfinance
import numpy as np #pip install numpy

#Creo el primer simbolo

symbol = 'AAPL'
ticker = yf.Ticker(symbol)
start_date = '2021-01-01'
end_date = '2022-09-01'
df = ticker.history(start = start_date, end = end_date)
df['Asset']='Apple'

#Creo el segundo simbolo

symbol2 = 'AMZN'
ticker2 = yf.Ticker(symbol2)
df2 = ticker2.history(start = start_date, end = end_date)
df2['Asset']='AMAZON'

#Creo el tercer simbolo

symbol3 = 'BAC'
ticker3 = yf.Ticker(symbol3)
df3 = ticker3.history(start = start_date, end = end_date)
df3['Asset']='Bank of America'

#Creo el cuarto simbolo

symbol4 = 'C'
ticker4 = yf.Ticker(symbol4)
df4 = ticker4.history(start = start_date, end = end_date)
df4['Asset']='Citygroup'

#Creo el quinto simbolo

symbol5 = 'GE'
ticker5 = yf.Ticker(symbol5)
df5 = ticker5.history(start = start_date, end = end_date)
df5['Asset']='General Electric'

#Creo el sexto simbolo

symbol6 = 'CL=F'
ticker6 = yf.Ticker(symbol6)
df6 = ticker6.history(start = start_date, end = end_date)
df6['Asset']='WTI'

#Creo el septimo simbolo

symbol7 = 'COP=X'
ticker7 = yf.Ticker(symbol7)
df7 = ticker7.history(start = start_date, end = end_date)
df7['Asset']='USD/COP'

Portafolio = pd.concat([df,df2,df3,df4,df5,df6,df7])
Portafolio.to_excel('Portafolio Inversion.xlsx')
Portafolio
```
